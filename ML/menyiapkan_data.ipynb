```python
import overpy
import time

api = overpy.Overpass()

# List nama kota/provinsi yang mau di-scrape
cities = [
    "Riau",
    "Jambi",
    "Bali",
    "Papua",
    "Kalimantan Tengah",
    "Sulawesi Utara"
]

def fetch_places(city_name, limit=20):
    query = f"""
    [out:json][timeout:25];
    area["name"="{city_name}"]["admin_level"="4"]->.searchArea;

    (
      node["tourism"="attraction"](area.searchArea);
      way["tourism"="attraction"](area.searchArea);
      relation["tourism"="attraction"](area.searchArea);

      node["tourism"="museum"](area.searchArea);
      way["tourism"="museum"](area.searchArea);
      relation["tourism"="museum"](area.searchArea);

      node["leisure"="park"](area.searchArea);
      way["leisure"="park"](area.searchArea);
      relation["leisure"="park"](area.searchArea);

      node["historic"](area.searchArea);
      way["historic"](area.searchArea);
      relation["historic"](area.searchArea);
    );

    out center {limit};
    """

    try:
        result = api.query(query)
    except Exception as e:
        print(f"Error querying {city_name}: {e}")
        return []

    places = []

    for node in result.nodes:
        name = node.tags.get("name")
        if not name:
            continue
        category = node.tags.get("tourism") or node.tags.get("leisure") or node.tags.get("historic") or "N/A"
        places.append({
            "City": city_name,
            "Place_Name": name,
            "Category": category,
            "Lat": node.lat,
            "Long": node.lon
        })

    for way in result.ways:
        name = way.tags.get("name")
        if not name:
            continue
        category = way.tags.get("tourism") or way.tags.get("leisure") or way.tags.get("historic") or "N/A"
        lat = getattr(way, "center_lat", None)
        lon = getattr(way, "center_lon", None)
        if lat is None or lon is None:
            continue
        places.append({
            "City": city_name,
            "Place_Name": name,
            "Category": category,
            "Lat": lat,
            "Long": lon
        })

    return places

all_places = []

for city in cities:
    print(f"Scraping kota: {city}")
    city_places = fetch_places(city, limit=20)
    all_places.extend(city_places)
    time.sleep(2)

# Print hasil (atau bisa simpan ke CSV, DB, dll)
for place in all_places:
    print(place)
```

```python
import pandas as pd
import requests
import time

def get_wikipedia_summary_id(place_name):
    """
    Ambil ringkasan dari Wikipedia bahasa Indonesia berdasarkan nama tempat.
    """
    url = f"https://id.wikipedia.org/api/rest_v1/page/summary/{place_name.replace(' ', '_')}"
    try:
        res = requests.get(url)
        if res.status_code == 200:
            data = res.json()
            if 'extract' in data:
                return data['extract']
            else:
                return None
        else:
            return None
    except Exception as e:
        return None

# Load CSV
df = pd.read_csv("/content/wisata_indonesia_osm.csv")

# Cek ada kolom Description belum, kalau belum buat
if 'Description' not in df.columns:
    df['Description'] = None

# Loop tiap tempat wisata, update kolom Description
for i, row in df.iterrows():
    place = row['Place_Name']
    print(f"Mencari deskripsi: {place} ({i+1}/{len(df)})")

    desc = get_wikipedia_summary_id(place)
    if desc:
        df.at[i, 'Description'] = desc
        print(f"Deskripsi ditemukan untuk {place}")
    else:
        print(f"Deskripsi tidak ditemukan untuk {place}")

    time.sleep(1)  # Untuk menghindari limit request API Wikipedia

# Simpan CSV hasil update
df.to_csv("wisata_indonesia_osm_updated.csv", index=False)
print("Selesai update deskripsi ke CSV.")
```

```python
# Instalasi library yang diperlukan
# !pip install requests beautifulsoup4 googlemaps selenium pandas numpy

import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import re

def scrape_wikipedia_description(place_name):
    """
    Fungsi sederhana untuk mendapatkan deskripsi dari Wikipedia Indonesia
    """
    try:
        # Bersihkan nama tempat untuk URL
        clean_name = place_name.replace(" ", "_")

        # Coba akses langsung halaman Wikipedia
        wiki_url = f"https://id.wikipedia.org/wiki/{clean_name}"
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        }

        response = requests.get(wiki_url, headers=headers)

        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')

            # Cari paragraf pertama
            content_div = soup.find('div', {'id': 'mw-content-text'})
            if content_div:
                paragraphs = content_div.find_all('p')
                for p in paragraphs:
                    text = p.get_text(strip=True)
                    if len(text) > 50:  # Minimal 50 karakter
                        return text[:500]  # Batasi 500 karakter

        # Jika tidak berhasil, coba menggunakan API pencarian
        search_url = "https://id.wikipedia.org/w/api.php"
        params = {
            'action': 'query',
            'format': 'json',
            'list': 'search',
            'srsearch': place_name,
            'srlimit': 1
        }

        response = requests.get(search_url, params=params)
        if response.status_code == 200:
            data = response.json()
            if data['query']['search']:
                title = data['query']['search'][0]['title']
                snippet = data['query']['search'][0]['snippet']
                # Bersihkan HTML tags dari snippet
                clean_snippet = re.sub('<.*?>', '', snippet)
                return clean_snippet

    except Exception as e:
        print(f"Error scraping Wikipedia for {place_name}: {e}")

    return None

def generate_simple_description(place_name, category, city):
    """
    Generate deskripsi sederhana berdasarkan kategori dan lokasi
    """
    category_descriptions = {
        'park': f"{place_name} adalah sebuah taman yang terletak di {city}. Tempat ini menyediakan ruang hijau untuk rekreasi dan bersantai bagi masyarakat.",
        'memorial': f"{place_name} adalah monumen peringatan yang berlokasi di {city}. Tempat ini memiliki nilai sejarah dan budaya yang penting.",
        'monument': f"{place_name} adalah sebuah monumen bersejarah di {city}. Bangunan ini menjadi landmark dan memiliki nilai historis.",
        'attraction': f"{place_name} adalah objek wisata menarik yang terletak di {city}. Tempat ini menawarkan pengalaman unik bagi pengunjung.",
        'museum': f"{place_name} adalah museum yang berlokasi di {city}. Tempat ini menyimpan koleksi dan informasi penting tentang sejarah dan budaya.",
        'temple': f"{place_name} adalah tempat ibadah yang terletak di {city}. Bangunan ini memiliki nilai spiritual dan arsitektur yang menarik.",
        'beach': f"{place_name} adalah pantai yang terletak di {city}. Tempat ini menawarkan pemandangan laut dan aktivitas wisata bahari."
    }

    return category_descriptions.get(category.lower(),
        f"{place_name} adalah tempat wisata yang terletak di {city}. Destinasi ini menawarkan pengalaman menarik bagi para pengunjung.")

# Fungsi utama untuk memproses dataframe
def enhance_tourism_data(df_path, output_path):
    """
    Fungsi utama untuk melengkapi data wisata
    """
    # Load data
    df = pd.read_csv(df_path)

    print(f"Loaded {len(df)} records")

    # Tambah kolom Rating jika belum ada
    if 'Rating' not in df.columns:
        df['Rating'] = None

    # Counter untuk tracking
    updated_count = 0

    for idx, row in df.iterrows():
        place_name = row['Place_Name']
        city = row['City']
        category = row['Category']
        current_desc = row['Description']

        print(f"Processing {idx+1}/{len(df)}: {place_name}")

        # Skip jika sudah ada deskripsi yang cukup panjang
        if pd.notna(current_desc) and len(str(current_desc)) > 50:
            print(f"  -> Already has description")
            continue

        # Coba dapatkan dari Wikipedia
        wiki_desc = scrape_wikipedia_description(place_name)

        if wiki_desc:
            df.at[idx, 'Description'] = wiki_desc
            print(f"  -> Updated from Wikipedia")
            updated_count += 1
        else:
            # Jika tidak ada di Wikipedia, buat deskripsi sederhana
            simple_desc = generate_simple_description(place_name, category, city)
            df.at[idx, 'Description'] = simple_desc
            print(f"  -> Generated simple description")
            updated_count += 1

        # Rating dummy (bisa diganti dengan scraping nyata)
        if pd.isna(row['Rating']):
            df.at[idx, 'Rating'] = round(3.5 + (hash(place_name) % 15) / 10, 1)  # Rating 3.5-5.0

        # Delay untuk menghindari rate limiting
        time.sleep(0.5)

        # Save progress setiap 50 record
        if (idx + 1) % 50 == 0:
            df.to_csv(output_path, index=False)
            print(f"  -> Progress saved at {idx+1} records")

    # Save final result
    df.to_csv(output_path, index=False)

    print(f"\nCompleted!")
    print(f"Updated {updated_count} descriptions")
    print(f"Results saved to {output_path}")

    return df

# Contoh penggunaan
if __name__ == "__main__":
    # Ganti path sesuai dengan lokasi file Anda
    input_file = "/content/wisata_indonesia_osm_updated.csv"
    output_file = "/content/wisata_indonesia_enhanced.csv"

    # Proses data
    enhanced_df = enhance_tourism_data(input_file, output_file)

    # Tampilkan sample hasil
    print("\nSample results:")
    print(enhanced_df[['Place_Name', 'City', 'Description', 'Rating']].head())

    # Statistik
    print(f"\nStatistics:")
    print(f"Total records: {len(enhanced_df)}")
    print(f"Records with description: {enhanced_df['Description'].notna().sum()}")
    print(f"Records with rating: {enhanced_df['Rating'].notna().sum()}")
    print(f"Average rating: {enhanced_df['Rating'].mean():.2f}")
```

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
import re

class TourismDataMerger:
    def __init__(self):
        # Mapping kategori dari scraped data ke kategori target
        self.category_mapping = {
            'park': 'Taman Hiburan',
            'memorial': 'Budaya',
            'monument': 'Budaya',
            'attraction': 'Wisata Alam',
            'museum': 'Wisata Budaya',
            'boundary_stone': 'Budaya',
            'castle': 'Wisata Budaya',
            'wayside_cross': 'Tempat Ibadah',
            'archaeological_site': 'Wisata Budaya',
            'ruins': 'Wisata Budaya',
            'manor': 'Wisata Budaya',
            'yes': 'Wisata Alam'
        }

        self.target_categories = [
            'Wisata Alam', 'Wisata Petualangan', 'Taman Hiburan',
            'Cagar Alam', 'Budaya', 'Wisata Hiburan', 'Wisata Kuliner',
            'Pusat Perbelanjaan', 'Tempat Ibadah', 'Wisata Budaya', 'Bahari'
        ]

    def clean_place_name(self, name):
        """Membersihkan nama tempat untuk matching yang lebih baik"""
        if pd.isna(name):
            return ""
        cleaned = str(name).lower()
        cleaned = re.sub(r'[^\w\s]', ' ', cleaned)
        cleaned = re.sub(r'\s+', ' ', cleaned).strip()
        return cleaned

    def calculate_distance(self, lat1, lon1, lat2, lon2):
        """Menghitung jarak antara dua koordinat (KM)"""
        try:
            return geodesic((lat1, lon1), (lat2, lon2)).kilometers
        except:
            return float('inf')

    def find_matches(self, df1, df2, distance_threshold=5.0, name_similarity_threshold=0.7):
        """
        Mencari kecocokan antara dua dataset berdasarkan nama dan jarak.
        Returns list of dict: { idx1, idx2, score, distance, name_similarity }
        """
        matches = []
        for idx1, row1 in df1.iterrows():
            name1 = self.clean_place_name(row1['Place_Name'])
            lat1, lon1 = row1['Lat'], row1['Long']
            best_match, best_score = None, 0.0

            for idx2, row2 in df2.iterrows():
                name2 = self.clean_place_name(row2['Place_Name'])
                lat2, lon2 = row2['Lat'], row2['Long']

                # Jaccard similarity sederhana
                words1, words2 = set(name1.split()), set(name2.split())
                if not words1 or not words2:
                    name_similarity = 0.0
                else:
                    intersection = len(words1 & words2)
                    union = len(words1 | words2)
                    name_similarity = intersection / union if union else 0.0

                distance = self.calculate_distance(lat1, lon1, lat2, lon2)

                if distance <= distance_threshold and name_similarity >= name_similarity_threshold:
                    score = name_similarity * 0.7 + (1 - min(distance / distance_threshold, 1)) * 0.3
                    if score > best_score:
                        best_match = {
                            'idx1': idx1,
                            'idx2': idx2,
                            'score': score,
                            'distance': distance,
                            'name_similarity': name_similarity
                        }
                        best_score = score

            if best_match:
                matches.append(best_match)

        return matches

    def map_category(self, original_category):
        """Mapping kategori scraped  kategori target."""
        if pd.isna(original_category):
            return 'Wisata Alam'
        return self.category_mapping.get(str(original_category).lower(), 'Wisata Alam')

    def merge_datasets(self, primary_df, secondary_df,
                       distance_threshold=5.0, name_similarity_threshold=0.5):
        """
        Menggabungkan dua DataFrame:
        - primary_df (DataFrame inmemory)
        - secondary_df (DataFrame inmemory)
        Prioritaskan primary_df: hanya update kolom kosong,
        tambahkan record baru jika belum ada.
        """
        print("Starting dataset merge...")
        print(f"Primary dataset: {len(primary_df)} records")
        print(f"Secondary dataset: {len(secondary_df)} records")

        # 1) Copy primary_df  merged_df
        merged_df = primary_df.copy()

        # 2) Pastikan kolom dasar ada di merged_df
        required_columns = ['Place_Id', 'Place_Name', 'Description', 'Category', 'City',
                            'Price', 'Rating', 'Time_Minutes', 'Lat', 'Long']
        for col in required_columns:
            if col not in merged_df.columns:
                if col == 'Place_Id':
                    merged_df[col] = range(1, len(merged_df) + 1)
                elif col == 'Price':
                    merged_df[col] = 0
                elif col == 'Time_Minutes':
                    merged_df[col] = 60
                else:
                    merged_df[col] = np.nan

        # 3) Cari pasangan match antara primary_df dan secondary_df
        matches = self.find_matches(
            merged_df, secondary_df,
            distance_threshold, name_similarity_threshold
        )
        print(f"Found {len(matches)} potential matches")

        used_secondary_indices = set()
        updated_count = 0

        # 4) Update kolom kosong di merged_df berdasarkan hasil matching
        for match in matches:
            idx1, idx2 = match['idx1'], match['idx2']
            if idx2 in used_secondary_indices:
                continue

            # Ambil baris dari merged_df & secondary_df
            primary_row = merged_df.loc[idx1]
            secondary_row = secondary_df.loc[idx2]
            updates_made = []

            # a) Update Description jika kosong atau terlalu pendek
            if (pd.isna(primary_row['Description']) or
                len(str(primary_row['Description'])) < 20):
                desc2 = secondary_row.get('Description', "")
                if pd.notna(desc2) and len(str(desc2)) >= 20:
                    merged_df.at[idx1, 'Description'] = desc2
                    updates_made.append('Description')

            # b) Update Rating jika kosong atau 0
            raw_rating = secondary_row.get('Rating')
            if isinstance(raw_rating, str):
                raw_rating = raw_rating.strip().lower()
                if raw_rating in ['none', 'nan', '']:
                    raw_rating = np.nan
                else:
                    try:
                        raw_rating = float(raw_rating)
                    except:
                        raw_rating = np.nan
            raw_rating = pd.to_numeric(raw_rating, errors='coerce')

            if (pd.isna(primary_row['Rating']) or primary_row['Rating'] == 0) and pd.notna(raw_rating):
                merged_df.at[idx1, 'Rating'] = raw_rating
                updates_made.append('Rating')

            # c) Update Category jika kosong
            if pd.isna(primary_row['Category']):
                merged_df.at[idx1, 'Category'] = self.map_category(secondary_row.get('Category'))
                updates_made.append('Category')

            if updates_made:
                updated_count += 1
                print(f"Updated row {idx1}: {', '.join(updates_made)} - {primary_row['Place_Name']}")

            used_secondary_indices.add(idx2)

        # 5) Tambahkan record baru (yang tidak match) dari secondary_df
        print("Adding new records from secondary dataset...")
        next_id = merged_df['Place_Id'].max() + 1 if not merged_df['Place_Id'].isna().all() else 1
        new_records = []

        for idx, row in secondary_df.iterrows():
            if idx in used_secondary_indices:
                continue

            raw_rating = row.get('Rating')
            if isinstance(raw_rating, str):
                raw_rating = raw_rating.strip().lower()
                if raw_rating in ['none', 'nan', '']:
                    raw_rating = np.nan
                else:
                    try:
                        raw_rating = float(raw_rating)
                    except:
                        raw_rating = np.nan
            raw_rating = pd.to_numeric(raw_rating, errors='coerce')

            new_records.append({
                'Place_Id': next_id,
                'Place_Name': row.get('Place_Name', ""),
                'Description': row.get('Description', ""),
                'Category': self.map_category(row.get('Category')),
                'City': row.get('City', ""),
                'Price': 0,
                'Rating': raw_rating,
                'Time_Minutes': 60,
                'Lat': row.get('Lat', np.nan),
                'Long': row.get('Long', np.nan)
            })
            next_id += 1

        if new_records:
            new_df = pd.DataFrame(new_records)
            merged_df = pd.concat([merged_df, new_df], ignore_index=True)
            print(f"Added {len(new_records)} new records")

        # 6) Bersihkan & standardisasi data
        merged_df = self.clean_merged_data(merged_df)

        print(f"\nMerge completed!")
        print(f"Final dataset: {len(merged_df)} records")
        print(f"Updated existing records: {updated_count}")
        added_count = len(merged_df) - len(primary_df)
        print(f"Actual new records added: {added_count}\n")

        return merged_df

    def clean_merged_data(self, df):
        """Membersihkan & standardisasi semua baris setelah merge"""
        # a) Standar kategori
        df['Category'] = df['Category'].fillna('Wisata Alam')
        df['Category'] = df['Category'].apply(
            lambda x: x if x in self.target_categories else 'Wisata Alam'
        )

        # b) Bersihkan Rating
        df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce').clip(0, 5).fillna(0)

        # c) Bersihkan Price
        df['Price'] = pd.to_numeric(df['Price'], errors='coerce').fillna(0).clip(0)

        # d) Bersihkan Time_Minutes
        df['Time_Minutes'] = pd.to_numeric(df['Time_Minutes'], errors='coerce').fillna(60).clip(15, 480)

        # e) Bersihkan Lat/Long
        df['Lat'] = pd.to_numeric(df['Lat'], errors='coerce')
        df['Long'] = pd.to_numeric(df['Long'], errors='coerce')

        before = len(df)
        df = df.dropna(subset=['Lat', 'Long']).reset_index(drop=True)
        if before != len(df):
            print(f"Removed {before - len(df)} records with invalid coordinates")

        return df

def merge_tourism_datasets(primary_df, secondary_df,
                           output_path=None,
                           distance_threshold=10.0,
                           name_similarity_threshold=0.3):
    """
    Menggabungkan dua DataFrame in-memory:
    - primary_df: DataFrame (combined_tourism_df_cleaned.copy())
    - secondary_df: DataFrame (pd.read_csv("wisata_indonesia_enhanced.csv"))
    Jika output_path diberikan (string), maka hasil juga disimpan ke CSV.
    """
    merger = TourismDataMerger()
    merged_df = merger.merge_datasets(
        primary_df.copy(),
        secondary_df.copy(),
        distance_threshold,
        name_similarity_threshold
    )

    if output_path:
        merged_df.to_csv(output_path, index=False)
        print(f"Merged dataset saved to {output_path}")

    print("\n=== SAMPLE MERGED DATA (first 10 rows) ===")
    print(merged_df[['Place_Name', 'Category', 'City', 'Description', 'Rating', 'Lat', 'Long']].head(10))

    return merged_df

# =======================
# Contoh Pemanggilan Kode
# =======================
if __name__ == "__main__":
    pass
```

```python
import pandas as pd
import numpy as np

# Misal final_df sudah ada di memori
# Cek dulu berapa banyak Rating=0
print("Sebelum imputasi, jumlah Rating=0:", (final_df['Rating'] == 0).sum())

def impute_rating_by_city_category(df):
    df = df.copy()

    # 1. Tandai Rating=0 sebagai NaN agar bisa diimputasi
    df['Rating'] = df['Rating'].replace(0, np.nan)

    # 2. Hitung rata-rata per grup (City, Category) dan isi NaN dengan rata-rata itu
    df['Rating'] = df.groupby(['City', 'Category'])['Rating'] \
                     .transform(lambda grp: grp.fillna(grp.mean()))

    # 3. Jika setelah langkah (2) masih ada NaN (artinya grup (City,Category) semuanya NaN),
    #    isi dengan rata-rata per Category
    df['Rating'] = df.groupby('Category')['Rating'] \
                     .transform(lambda grp: grp.fillna(grp.mean()))

    # 4. Jika masih ada NaN (misalnya Category baru yang semua NaN), isi dengan rata-rata keseluruhan
    global_mean = df['Rating'].mean()
    df['Rating'] = df['Rating'].fillna(global_mean)

    # 5. Pastikan tetap dalam rentang 05
    df['Rating'] = df['Rating'].clip(0, 5)

    return df

# Terapkan fungsi imputasi
final_df = impute_rating_by_city_category(final_df)

# Tampilkan hasil
print("Setelah imputasi, jumlah Rating=0:", (final_df['Rating'] == 0).sum())
print("\nContoh beberapa baris pertama:")
print(final_df[['Place_Name', 'City', 'Category', 'Rating']].head(10))
```
